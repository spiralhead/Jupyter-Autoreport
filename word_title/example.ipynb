```python
import ipywidgets as widgets
from IPython.display import display
import ipynbname
import os
btn = widgets.Button(description='Convert Notebook')
output = widgets.Output()
display(btn, output)
@output.capture()
def btn_eventhandler(obj):
    output.clear_output()
    print('Определяем полный путь до блокнота')
    nb_fname = ipynbname.path()
    convert_command = f'''call conda activate {os.environ['CONDA_DEFAULT_ENV']}
call jupyter nbconvert --execute --config=cfg.py --to pdf --no-input {nb_fname}
call python final_merge.py
    '''
    with open('nbconvert.bat','w',encoding='utf8') as f:
        print('Создаем файл nbconvert.bat')
        f.write(convert_command)
    print('Запускаем nbconvert.bat')
    errorcode = os.system('nbconvert.bat')
    if errorcode == 0:
        print('Nbconvert успешно завершился!')
    else:
        print('ERROR--Конвертация завершилась с ошибкой!!!')
btn.on_click(btn_eventhandler)
```

# Эвольвента
Эвольвента окружности - траектория, описываемая любой точкой на прямой, которую перекатывают по окружности без скольжения
$$x = r_o (\cos\alpha+\alpha\sin\alpha)$$ 
$$y = r_o (\sin\alpha-\alpha\cos\alpha)$$
График функции:


```python
import numpy as np
import matplotlib.pyplot as plt

x = lambda r, a: r * (np.cos(a) + a * np.sin(a))
y = lambda r, a: r * (np.sin(a) - a * np.cos(a))
r, a = np.linspace(0, 10, 1000), np.linspace(0, 10, 1000)
plt.plot(x(r, a), y(r, a));
```
